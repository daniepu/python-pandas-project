## importing the data and create main tables
we got pandas and numpy which are two python library helped us arrange the data.<br>
the data imported from the csv files and the one of the two main table was created from there sales data files,<br>
the sales table contain the data of all other tables we need to create .


```python
import pandas as pd
import numpy as np
sales19 = pd.read_csv("Training_Dataset.xlsx - 2019 Sales.csv",parse_dates = ["Purchase Date"])
sales18 = pd.read_csv("Training_Dataset.xlsx - 2018 Sales.csv",parse_dates = ["Purchase Date"])
sales17 = pd.read_csv("Training_Dataset.xlsx - 2017 Sales.csv",parse_dates = ["Purchase Date"])
monthTarget = pd.read_csv("Training_Dataset.xlsx - Monthly Target.csv")
yearTarget = pd.read_csv("Training_Dataset.xlsx - Yearly Target.csv")
sales = pd.concat([sales17,sales18,sales19],ignore_index=True,keys = [sales17,sales18,sales19])
sales.dropna(how = "all")
sales.fillna("None", inplace = True)
sales['Cost Per Unit'] = sales['Cost Per Unit'].replace({',':''},regex=True)
sales['Cost Per Unit'] = sales['Cost Per Unit'].astype(float)
sales.to_csv(r'sales17-19.csv',index = False)
sales.head()
```

## seconed main table
the targets table containes to tables of month and year targets ,<br>
they cleaned and converted into the seconed main table that containes all data<br>
we need about the targets for each location in 2019 and stored in csv file<br>

```python
def monthtargetcalculate(row):
    totaltarget = row[2]
    precent = row[4]
    return totaltarget*precent


def dateid(row):
    if row[3] < 9:
        s = str(row[0]) + '0' + str(row[3]) + '01'
    else:
        s = str(row[0]) + str(row[3]) + '01'
    return s

Targets = yearTarget.merge(monthTarget,how="left")
#Targets["Month Target"] = monthTarget['%  Yearly Sales Target ']*yearTarget['Sales Target']
Targets["Month Target"] = Targets.apply(monthtargetcalculate,axis="columns")
Targets["Date ID"] = Targets.apply(dateid,axis="columns")
Targets = Targets.loc[:,["Location ID","Year","Month","Month Target","Sales Target",'%  Yearly Sales Target ',"Date ID"]]
Targets.to_csv(r'Targets.csv',index = False)
Targets.head()
```

## Salesman csv file creation

```python
salesman = sales.loc[:,["Sales Person ID","Salesperson Name"]]
salesman = salesman.groupby(["Sales Person ID","Salesperson Name"]).count().reset_index()
salesman.to_csv(r'salesman.csv',index = False)
salesman.head()
```

## Location csv file creation

```python
l = sales.loc[:,['Location ID'
      ,'County'
      ,'State'
      ,'Latitude'
      ,'Longitude'
      ,'Area Population'
      ,'Area Households'
      ,'Area Median Income'
      ,'City Name']]
Location = l.groupby(['Location ID'
      ,'County'
      ,'State'
      ,'Latitude'
      ,'Longitude'
      ,'Area Population'
      ,'Area Households'
      ,'Area Median Income'
      ,'City Name']).count().reset_index()
Location.to_csv(r'Location.csv',index = False)
Location.head()
```

## Customers csv file creation

```python
Customers = sales.loc[:,['Customer ID'
      ,'Customer Name'
      ,'Age'
      ,'Gender']]
Customers = Customers.groupby(['Customer ID'
      ,'Customer Name'
      ,'Age'
      ,'Gender']).count().reset_index()
Customers.to_csv(r'Customers.csv',index = False)
Customers.head()
```

## Products csv file creation

```python
products = sales.loc[:,['Product ID'
      ,'Product Name'
      ,'Price'
      ,'Cost Per Unit'
      ,'Category ID'
      ,'Category Name']]
products = products.groupby(['Product ID'
      ,'Product Name'
      ,'Price'
      ,'Cost Per Unit'
      ,'Category ID'
      ,'Category Name']).count().reset_index()
products.to_csv(r'products.csv',index = False)
products.head()
```

## Orders csv file creation

```python
orders = sales.loc[:,['Order ID'
      ,'Product ID'
      ,'Location ID'
      ,'Sales Person ID'
      ,'Customer ID'
      ,'Quantity'
      ,'Price'
      ,'Cost Per Unit'
      ,'Taxes'
      ,'Purchase Date']]
orders["revenue"] = orders['Quantity']*(orders['Price'] - orders['Cost Per Unit'])
orders["totalsales"] = orders['Price']*orders['Quantity']
orders['Purchase Date'] = orders['Purchase Date'].astype(str).replace({'-':''},regex = True)
orders['Purchase Date'] =  orders['Purchase Date'].astype(int)
orders.to_csv(r'orders.csv',index = False)
orders.head()
```

## Dates csv file creation

```python
def month_indc(row):
    if row[0].year == 2019:
        return 1
    return 0

def year_indc(row):
    if row[0].month == 3:
        return 1
    return 0

times = pd.date_range(start='1/1/2017',end = '31/12/2019',freq = 'D')
dates = pd.DataFrame({"fulldate" : times})
dates["DateID"] =  dates["fulldate"].astype(str).replace({'-':''},regex = True).astype(int)
dates["week day"] = pd.DatetimeIndex(dates['fulldate']).weekday
dates["Month ID"] = pd.DatetimeIndex(dates['fulldate']).month
dates["Month Discription"] = pd.DatetimeIndex(dates['fulldate']).month_name()
dates["Year ID"] = pd.DatetimeIndex(dates['fulldate']).year
dates["Quarter ID"] = pd.DatetimeIndex(dates['fulldate']).quarter
dates["month_ind"] = dates.apply(month_indc,axis="columns")
dates["year_ind"] = dates.apply(year_indc,axis="columns")
dates.to_csv(r'dates.csv',index = False)
dates.head()
```
